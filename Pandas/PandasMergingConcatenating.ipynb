```python
#Example data frames 
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']})
                     
```

```python
#MERGING 
#The merge function allows you to merge DataFrames together using similar logic as merging SQL tables together. 
#EXAMPLE:
#pass in the left df and right df merge them in inner 
# when merging merge on a key column on the key coloum wehre left and right df has K1, K2 .....
pd.merge(left,right,how='inner',on='key')
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A0', 'A1', 'A2'],
                     'B1': ['B0', 'B0', 'B1', 'B2']})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K0'],
                     'key2': ['K0', 'K0', 'K0', 'K0'],
                     'C': ['C0', 'C0', 'C1', 'C2'],
                     'D1': ['D0', 'D0', 'D1', 'D2']})
```

```python
pd.merge(left, right, on=['key1', 'key2']) 
# if you want to merge key 1 with key 2
#list ofkeys if you want to merge several keys
```

```python
pd.merge(left, right, how='outer', on=['key1', 'key2']) 
#outer right is SQL syntax left is usually inner join 
```

```python
pd.merge(left, right, how='left', on=['key1', 'key2']) 
```

```python
pd.merge(left, right, how='right', on=['key1', 'key2']) 
```

```python
#Joining is a method for combing the columns of two potentially differently-indexed Dataframes into a single result Dataframe.
left = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']},
                     index=['K0', 'K0', 'K1', 'K2'])

right = pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']},
                     index=['K0', 'K0', 'K2', 'K3'])
```

```python
left.join(right)
```

```python
left.join(right, how='outer')
```
