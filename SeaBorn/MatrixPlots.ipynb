```python
import seaborn as sns
import numpy as np
sns.set_style('darkgrid')
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
#heat maps which is our primarly way of showing some sort of matrix plot in order
#matrix plot matrix form where the index name and the column name match up so that the cell
#vulue that actually inciated both of their names
```

```python
tc = tips.corr()
#matrix form in order for the heat map
#heat map color the value based on a gradients scale 
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
# now he have our data in matrix forms
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidths=1)
#base in the heat map the recent year and summer time you get more flights in US data
#line color and linewidt makes it easier to read if you want more seperations
```

```python
#cluster map uses hierachal clustering to produce a cluster versoion of this heatmap
#you may have to wait for clustering algorithms to better understand what 
# actually going on the mathematics beheind these methods
sns.clustermap(fp,cmap='magma',standard_scale=1)
```
