```python
import seaborn as sns
import numpy as np
sns.set_style('darkgrid')
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
#BAR PLOT is a basic plot that aggregate the categorical data based on some function
#estimator is a funciton that is a agragrate function that is the average or mean or standard deviation per example stadern
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
# this shows the mean of the total bill ber categorical value
```

```python
# a count plot is similar to a bar plot except that the estimator is explicitly counting the number of occurences
sns.countplot(x='sex',data=tips)
```

```python
# a box plot are used to show the distrubution of quantative data in a way
# that hopefully facilitates comparisons between the variabls
# the box plots shows teh quartiles of the dataset
#the whiskers shows the rest of the distribution
#the point outside the whiskers are outliers
#hue indicator compare distrubution of the total_bill by changing the hue
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
#violin plots
#violin plot allows all of teh components that correspond to actually data point and essentially showing the 
#underlying distrubion and releave teh datapoint themselve
#but the cost comes that they are hard to read
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
# stripplot 
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)

# swarm plot is like a combination of a strip plot and violin plot
# dont use swarm plots with big data sets
```

```python
#you cna combine plots
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```
