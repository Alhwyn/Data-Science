```python
# Grid aoutmate subplot based on feature on our data
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
#this dataset is just a bunch of measurements of irises flowers
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
b = sns.PairGrid(iris)
b.map_diag(sns.displot)
b.map_upper(plt.scatter)
b.map_lower(sns.kdeplot)
```

```python
#Fasset Grids 
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```
