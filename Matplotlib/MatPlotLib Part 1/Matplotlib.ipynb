```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL Method
plt.plot(x,y, 'r-')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OBJECT ORIENTED METHOD OO
```

```python
fig = plt.figure() # this figure object ix a blank canvas

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X - axis')
axes.set_xlabel('X - axis')
axes.set_ylabel('Y - axis')
axes.set_title('x^2')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.05,0.9,0.9])
axes2 = fig.add_axes([0.58, 0.15,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('Inverse Plot')
axes1.set_title('Parent Function')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')


axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```
