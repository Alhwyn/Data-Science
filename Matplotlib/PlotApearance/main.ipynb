```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
# you cant put in a RGB HEX CODE BY PUTTING # IN THE BEGGING
#alpha changes the transparency
# linestyle changes the line
# -- makes a dash line
# -. makes a dash point line
# .. makes a pointed line
# : makes smaller dots
# steps for stairs
# markers plots the pointso the graph
#you can change the size of teh marker by markersize
axes.plot(x,y,color='blue',lw=1,ls='--')

axes.set_xlim([0,1])

axes.set_ylim([0,2])
```
